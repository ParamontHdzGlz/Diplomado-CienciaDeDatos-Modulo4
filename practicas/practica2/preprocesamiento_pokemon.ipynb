```python
from glob import glob                                                           
from PIL import Image
from pathlib import Path
import os 
import cv2
import numpy as np 

import shutil
import pandas as pd
import errno

from sklearn.model_selection import train_test_split
from numpy import array

path_dir = 'dataset/images/'
train_path = Path(path_dir)
jpgs = list(train_path.glob('*.jpg'))
len(jpgs)
```

### transformar imagenes jpg a png

```python
for j in jpgs:
    j = str(j)
    img = cv2.imread(j)
    
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    v = hsv[:,:,2]
    th, threshed = cv2.threshold(v, 100, 255, cv2.THRESH_OTSU|cv2.THRESH_BINARY_INV)
    threshed[-1] = 255

    cnts = cv2.findContours(threshed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)[-2]

    mask = np.zeros_like(threshed)
    cv2.drawContours(mask, cnts, -1, (255, 0, 0), -1, cv2.LINE_AA)
    mask = cv2.erode(mask, np.ones((3,3), np.int32), iterations=1)

    png = np.dstack((img, mask))
    
    cv2.imwrite(j[:-3] + 'png', png)
    os.remove(j)
    print(str(j),"replaced successfully!")
```

```python
df = pd.read_csv("dataset/pokemon.csv")
df
```

```python
old=r"dataset/images"### Ruta donde se encuntran las imagenes 
new_train=r"./Práctica 2/train" 
new_test=r"./Práctica 2/test" 
```

```python
def mover_pokemon(pokemon,old,new):
    try:
        shutil.move(os.path.join(old+"/"+pokemon), os.path.join(new+"/"+pokemon))
    except OSError as e:
        if e.errno != errno.ENOENT:
            raise

def crear_carpetas(ruta,carpetas):
    
    try:
        os.mkdir(ruta)
    except OSError as e:
        if e.errno != errno.EEXIST:
            raise
    
    for x in carpetas:
        try:
            os.mkdir(ruta+"/"+x)
        except OSError as e:
            if e.errno != errno.EEXIST:
                raise
        
```

## Filtramos los Tipos de Pokemon

```python
tipos=list(set(df["Type1"]))
tipos
```

## Creamos las carpetas 

```python
crear_carpetas(new_train,tipos)
crear_carpetas(new_test,tipos)
```
